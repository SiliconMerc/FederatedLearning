```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =[0,5,10,15,20]
boys=np.array([0.010177527673086394,0.03576845630516899,0.12341125620982685,0.13592181129202546,0.1422611130941367])
girls=-boys
w=0.4
x1=np.arange(len(boys))
x2=x1+w
plt.bar(x1,boys,w,label="0<=n<=4")
plt.bar(x2,girls,w,label="5<=n<=9")
plt.xlabel("(u/-u)")
plt.ylabel("Mn")
plt.xticks((x1+x2)/2,x)
plt.legend()
plt.plot()
plt.show()
```
